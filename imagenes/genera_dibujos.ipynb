```python
#export
from diffusers import StableDiffusionXLPipeline, FluxPipeline

import torch
import os
```

```python
#export

def genera_imagen(prompt, root, file_output, height=1200, width=1800, num_inference_steps=40, guidance_scale=7.5, max_length=400, model_id="stabilityai/stable-diffusion-xl-base-1.0"):
    method="genera_imagen"

    # Carga el tokenizador con un valor más alto (por ejemplo, 150)
    #tokenizer = CLIPTokenizer.from_pretrained(model_id, model_max_length=max_length)
    if model_id=="stabilityai/stable-diffusion-xl-base-1.0":
        pipe = StableDiffusionXLPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    elif model_id=="black-forest-labs/FLUX.1-schnell":
        pipe = FluxPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
        pipe.enable_model_cpu_offload()
        
    
    file_out_path=os.path.join(root, f"{file_output}.png")

    pipe.to("mps")
    image = pipe(
        prompt=prompt, 
        height=height, 
        width=width, 
        num_inference_steps=num_inference_steps, 
        guidance_scale=guidance_scale,
        max_token_length=max_length
    ).images[0]
    image.save(file_out_path)
```

```python
#Portada 

prompt='''
Estilo de cuento infantil. Ilustración mágica.
Centro abajo una hormiga debajo de un gran árbol, rodeada de un bosque encantado.
Es de noche. Fondo estrellado iluminado. 
Cielo y solo la silueta de dragón verde. 
Ambiente cálido y misterioso. 
Área despejada en la parte superior para el título. 
'''
root=""
file_output="portada"
genera_imagen(prompt, root, file_output)
```

```python
#Ilustración 1
prompt='''
Ant looking to an ant nest in a magical star night. Next to a very big magical tree with lots of lights, foliage, leaves and branches. 
In the upper part neutral color to write text.
'''
root=""
file_output="ilustracion_1_2"
genera_imagen(prompt, root, file_output, max_length=len(prompt) + 10)
```
